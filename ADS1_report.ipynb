```python
#Pandas is used to analyze data
import pandas as pd
#Matplotlib is a plotting library
import matplotlib.pyplot as plt
#Seaborn is a Python data visualization library
import seaborn as sns
%matpplotlib inline
```

```python
#Read the dataset
data = pd.read_csv('data.csv')
```

```python
data.head()
```

```python
#dispply Bar chart 
data = data.dropna(axis=1)
'''
Drop the column with all missing values (na, NAN, NaN)
NOTE: This drops the column Unnamed: 32 column
Get a count of the number of 'M' & 'B' cells'''

data['diagnosis'].value_counts()
#Visualize this count 

sns.countplot(data['diagnosis'],label="Count")

```

```python
# create Histogram for the given dataset
data.plot(kind = 'hist')
'''
kind hist will plot the Histogram and in this example we are ploting it for all the features oof dataset 
'''
plt.title("Histogram")
plt.show()
```

```python
#plot piechart
fig = plt.figure()
'''
The figure() function in pyplot module of matplotlib library is used to create a new figure. 
 ax , short for axes
 rc , defines a runtime configuration
 autopct, display the percent value 
'''
fig,ax = plt.subplots(figsize=(8,8))
plt.rcParams['font.size'] = 15.0
data['diagnosis'].value_counts(sort=False).plot(kind='pie',autopct='%1.0f%%')
```

```python
#Display the Scatterplot
g1 = data.loc[data.diagnosis==1,:]
'''
loc(), retrieve data values from a dataset
plt, plot the figure of size 10,7 
where on x axes we have diagnosis and  radius_mean on Y axes from the dataset
'''
plt.figure(figsize=(10,7))
sns.scatterplot(x = 'diagnosis', y = 'radius_mean', s = 70, data=data);
      
```
